Tsv similar to csv, hence just specify the delimiter=tab in load_datasets() method.

```python
from datasets import load_dataset

data_files = {"train": "drugsComTrain_raw.tsv", "test": "drugsComTest_raw.tsv"}
# \t is the tab character in Python
drug_dataset = load_dataset("csv", data_files=data_files, delimiter="\t")
```

```python
#To verify our dataset
drug_sample = drug_dataset["train"].shuffle(seed=42).select(range(1000))
# Peek at the first few examples
drug_sample[:3]
```

 To test the patient ID hypothesis for the Unnamed: 0 column, we can use the Dataset.unique() function to verify that the number of IDs matches the number of rows in each split.

```python
for split in drug_dataset.keys():
    assert len(drug_dataset[split]) == len(drug_dataset[split].unique("Unnamed: 0"))
```

```python
#Our hypothesis confiremed and hence we use the rename_column method to rename our column unnamed to patient_id.
drug_dataset = drug_dataset.rename_column(
    original_column_name="Unnamed: 0", new_column_name="patient_id"
)
drug_dataset
```

Now we try normalizing the condition labels, however we get an error because there are some None entries and we can't normalize to lowercase since the None entries are not strings!

```python
def lowercase_condition(example):
    return {"condition": example["condition"].lower()}


drug_dataset.map(lowercase_condition)
```

```python
#Rectifying the error by using the filter method with a condition to select entries that are not None by using a lambda function!
drug_dataset = drug_dataset.filter(lambda x: x["condition"] is not None)
```

```python
drug_dataset
```

With the None entries removed, we can normalize our condition column:

```python
drug_dataset = drug_dataset.map(lowercase_condition)
# Check that lowercasing worked
drug_dataset["train"]["condition"][:3]
```

Creating new columns to add the length of the reviews which can be single words or multiple paragraphs.

The compute_length() method will be mapped to the dataset(), to apply the compute_length() method to every other row in the dataset!

```python
def compute_review_length(example):
    return {"review_length": len(example["review"].split())}
```

```python
drug_dataset = drug_dataset.map(compute_review_length)
# Inspect the first training example
drug_dataset["train"][0]
```

```python
#sort() method to check what the extreme values wld look like!
drug_dataset["train"].sort("review_length")[:3]
```

To filter the reviews that are lesser than 30 words.

```python
drug_dataset = drug_dataset.filter(lambda x: x["review_length"] > 30)
print(drug_dataset.num_rows)
```

To remove the html character codes in our reviews using html module of Python.

```python
drug_dataset = drug_dataset.map(lambda x: {"review": html.unescape(x["review"])})
```

Speeding the batch processing using list comprehension method of map(), by specifying batched=True.

```python
new_drug_dataset = drug_dataset.map(
    lambda x: {"review": [html.unescape(o) for o in x["review"]]}, batched=True
)
```

Various uses of batched=True!!

```python
#using the fast method(batched=True)
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")


def tokenize_function(examples):
    return tokenizer(examples["review"], truncation=True)
```

```python
%time tokenized_dataset = drug_dataset.map(tokenize_function, batched=True)
```

As we can compare the above two time cells, batched=True, remarkably performs better!

Here are the results we obtained with and without batching, with a fast and a slow tokenizer:

batched=True	                10.8s	                  4min41s

batched=False	                59.2s	                  5min3s

Parallelization in Dataset.map().

```python
slow_tokenizer = AutoTokenizer.from_pretrained("bert-base-cased", use_fast=False)


def slow_tokenize_function(examples):
    return slow_tokenizer(examples["review"], truncation=True)


tokenized_dataset = drug_dataset.map(slow_tokenize_function, batched=True, num_proc=8)
```

The below code helps map without removing the excess columns!

```python
def tokenize_and_split(examples):
    result = tokenizer(
        examples["review"],
        truncation=True,
        max_length=128,
        return_overflowing_tokens=True,
    )
    # Extract mapping between new and old indices
    sample_map = result.pop("overflow_to_sample_mapping")
    for key, values in examples.items():
        result[key] = [values[i] for i in sample_map]
    return result
```

```python
tokenized_dataset = drug_dataset.map(tokenize_and_split, batched=True)
tokenized_dataset
```
